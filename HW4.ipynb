```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# (a)
np.random.seed(50)

n_obs = 20
n_features = 50
n_classes = 3

class1 = np.random.normal(loc=0, scale=1, size=(n_obs, n_features))
class2 = np.random.normal(loc=0, scale=1, size=(n_obs, n_features)) + 5
class3 = np.random.normal(loc=0, scale=1, size=(n_obs, n_features)) - 5

data = np.vstack([class1, class2, class3])
true_labels = np.array([0]*n_obs + [1]*n_obs + [2]*n_obs)

# (b)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data)

colors = ['red', 'orange', 'blue']
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.scatter(pca_result[true_labels == i, 0],
                pca_result[true_labels == i, 1],
                label=f'Class {i+1}', color=colors[i])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA')
plt.legend()
plt.show()
```
